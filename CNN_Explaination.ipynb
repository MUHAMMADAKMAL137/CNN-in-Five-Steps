```python
import numpy as np
from numpy import asarray
```

```python
data = [[3,3,2,1,0],
        [0,0,1,3,1],
        [3,1,2,2,3],
        [2,0,0,2,2],
        [2,0,0,0,1],]
        #the variable data is a 2D list representing a matrix with 5 rows and 5 columns, where each element contains integer values ranging from 0 to 3.
```

```python
data = np.asarray(data)
data =data.reshape(1,5,5,1)
#the variable data is converted to a NumPy array using np.asarray(), and then it is reshaped to a 4-dimensional array with a shape of (1, 5, 5, 1), representing a single grayscale image of size 5x5.
```

```python
kernel = [[[[0]], [[1]], [[2]]],
          [[[2]], [[2]], [[0]]],
          [[[0]], [[1]], [[2]]]]
          #the variable kernel is a 4-dimensional list representing a 3x3 filter with a depth of 1, used for convolutional operations in a deep learning model.
```

```python
weights = [asarray(kernel), asarray([0.0])]
#the variable weights is a list containing two elements: the first element is a NumPy array kernel, and the second element is a NumPy array with a single value 0.0, which could be used as weights in a neural network layer.
```

```python
from keras.models import Sequential
from keras.layers import Conv2D
```

```python
model = Sequential()
model.add(Conv2D(1,(3,3), input_shape=(5,5,1)))
model.summary()
#a sequential model is initialized, followed by adding a 2D convolutional layer with 1 filter and a kernel size of (3,3), taking input with a shape of (5,5,1), and finally, the model summary is displayed.
```

```python
model.set_weights(weights)
yhat= model.predict(data)
for r in range (yhat.shape[1]):
  print([yhat[0,r,c,0] for c in range(yhat.shape[2])])
  #the pre-defined weights are set to the model, then using the input data, predictions are made using the model, and the output predictions for each row are printed in a formatted way.
```

```python
from keras.layers.pooling.max_pooling2d import MaxPool2D
model = Sequential()
model.add(Conv2D(1,(3,3), padding = 'same',input_shape=(5,5,1)))
model.add(MaxPool2D((2,2)))
#a 2D max-pooling layer with a pool size of (2,2) is added to the model, reducing the spatial dimensions of the input data by a factor of 2 in both height and width.
model.summary()
model.set_weights(weights)
yhat= model.predict(data)
for r in range (yhat.shape[1]):
  print([yhat[0,r,c,0] for c in range(yhat.shape[2])])
```

```python
from keras.layers.pooling.average_pooling2d import AveragePooling2D
from keras.layers.pooling.max_pooling2d import MaxPool2D
model = Sequential()
model.add(Conv2D(1,(3,3), padding = 'same',input_shape=(5,5,1)))
model.add(AveragePooling2D((2,2)))
#a 2D Avg-pooling layer with a pool size of (2,2) is added to the model, reducing the spatial dimensions of the input data by a factor of 2 in both height and width.
model.summary()
model.set_weights(weights)
yhat= model.predict(data)
for r in range (yhat.shape[1]):
  print([yhat[0,r,c,0] for c in range(yhat.shape[2])])
```

```python
from keras.layers import Flatten
from keras.layers.pooling.max_pooling2d import MaxPool2D
model = Sequential()
model.add(Conv2D(1,(3,3), padding = 'same',input_shape=(5,5,1)))
model.add(MaxPool2D((2,2)))
model.add(Flatten())
# 2D convolutional layer with 1 filter and a kernel size of (3,3) is added to the model with 'same' padding and taking input with a shape of (5,5,1). Then, a 2D max-pooling layer with a pool size of (2,2) is added to the model, followed by a flatten layer to transform the multi-dimensional output into a one-dimensional vector.
model.summary()
```
